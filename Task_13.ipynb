```python
import sys

for line in sys.stdin:
    text1 = line.strip()

len_word = ['1', '21', '213', '3142', '51324', '614253', '5172436', '61825374', '816273549']
text1 = text1.split()

new_text = ''
for word in text1:
    flag = 0
    if word[-1] not in 'QWERTYUIOPLKJHGFDSAZXCVBNMqwertyuioplkjhgfdsazxcvbnm':
        x = word[-1]
        flag = 1
        word = word[:-1]
    text = ''
    for i in len_word:
        if len(i) == len(word):
            new_i = list(i)
            for idx in new_i:
                text = text + word[int(idx) - 1]
    new_text = new_text + ' ' + text
    if flag == 1:
        new_text = new_text + x
if new_text[0] == ' ':
    new_text = new_text[1:]

def cezar(text, k):
    alphabet = [chr(i) for i in range(65, 91)] + [chr(j) for j in range(97, 123)]
    moch = 26
    for i in range(len(text)):
        if text[i].isalpha():
            if text[i].isupper():
                print(alphabet[(alphabet.index(text[i]) + k) % moch], end = '')
            else:
                print(alphabet[(alphabet.index(text[i]) + k) % moch + moch], end='')
        else:
            print(text[i], end = '')

alphabet_big = [chr(i) for i in range(65, 91)]
alphabet_big = ''.join(alphabet_big)
alphabet_small = [chr(j) for j in range(97, 123)]
alphabet_small = ''.join(alphabet_small)

s = new_text.split()
for i in range(len(s)):
    if s[i][0] in alphabet_big:
        idx = alphabet_big.find(s[i][0])
    if s[i][0] in alphabet_small:
        idx = alphabet_small.find(s[i][0])
    
    cezar(s[i], idx + 1)
    print(' ', end='')
```
