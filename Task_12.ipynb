```python
import sys
text = []
for line in sys.stdin:
    text.append(line.strip())
output = []
for elem in text:
    if len(output) == 0:
        x = elem.split(':')
        y = x[1].split(',')
        for i in y:
            el = x[0] + '-' + i
            output.append(el)
    else:
        flag = 0
        x = elem.split(':')
        y = x[1].split(',')
        for out in output:
            idx = out.rfind('-')
            if out[idx+1:] == x[0]:
                flag = 1
                for i in y:
                    el = out + '-' + i
                    output.append(el)
                output.remove(out)
                continue
        if flag == 0:
            for i in y:
                el = x[0] + '-' + i
                output.append(el)
count = 0
for out in output:
    if out.count('-') > 4:
        count += 1
print(count)
```
