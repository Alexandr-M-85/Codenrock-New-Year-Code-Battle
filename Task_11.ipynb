```python
import sys
lis = []
for line in sys.stdin:
    lis.append(line.strip())
wilyam = lis[-1]
dolores = lis[-2]
lis = lis[:-2]
for i, row in enumerate(lis):
    if 'W' in row:
        wilyam_y = i
        wilyam_x = row.index('W')
    if 'D' in row:
        dolores_y = i
        dolores_x = row.index('D')

if wilyam == 'right':
    wilyam_x += 1
elif wilyam == 'left':
    wilyam_x -= 1
elif wilyam == 'up':
    wilyam_y -= 1
elif wilyam == 'down':
    wilyam_y += 1
    
if dolores == 'right':
    dolores_x += 1
elif dolores == 'left':
    dolores_x -= 1
elif dolores == 'up':
    dolores_y -= 1
elif dolores == 'down':
    dolores_y += 1

if abs(wilyam_x - dolores_x) == abs(wilyam_y - dolores_y):
    print('Yes')
else:
    print('No')
```
